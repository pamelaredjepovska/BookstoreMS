# [dbo].[UpdateAuthorID]

```sql
/****** Object:  StoredProcedure [dbo].[UpdateAuthorID]    Script Date: 07.7.2020 14:57:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[UpdateAuthorID]
	@oldID nvarchar(11),
	@newID nvarchar(11)
AS
BEGIN
	UPDATE BooksDB..Authors
	SET au_id=@newID
	WHERE au_id LIKE '%@oldID%'
END
GO
```
